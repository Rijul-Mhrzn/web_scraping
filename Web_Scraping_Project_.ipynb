```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.worldometers.info/gdp/nepal-gdp/'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
print(soup.prettify())
```

```python
first = soup.find_all('table')[1]
```

```python
print(first.prettify())
```

```python
second = soup.find('table', class_ = "table table-striped table-bordered table-hover table-condensed table-list")
```

```python
print(second.prettify())
```

```python
table = soup.find_all('table')[1]
print(table)
```

```python
gdp_titles = table.find_all('th')
```

```python
gdp_titles
```

```python
gdp_table_titles = [title.text.strip() for title in gdp_titles]
```

```python
print(gdp_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = gdp_table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
print(table.prettify())
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data

```

```python
df
```

```python
df.to_csv(r'/content/table.csv', index = False)
```
